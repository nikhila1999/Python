```
# using function to return large number from list
def largenumber(li):
  li.sort()
  return li[-1]
def genericlarge(li,n):
  li.sort()
  return li[-n]
li=[19,1,3,87,32,21,36]
print(largenumber(li))
print(genericlarge(li,3))

```

```
def smallnumber(li):
  li.sort()
  return li[0]
def genericsmall(li,n):
  li.sort()
  return li[n-1]
li=[19,1,3,87,32,21,36]
print(smallnumber(li))
print(genericsmall(li,3))
```

```
#given string is palindrome or not

def palindrome(str):
  if str==str[::-1]:
    return True
  else:
    return False
palindrome("mom")
```

#Conditional Statements
## If - else Statements
### Syntax:
#### if condition:
> Statements<br>
##### else:
> statements

```
# even or odd
n=int(input("enter a number:"))
if n%2==0:
  print("even number")
else:
  print("odd number")
```

```
# check given num is perfectly multiple of 3 and 5
n=int(input("enter a number:"))
if (n%3==0 and n%5==0):
  print("divisible")
else:
  print("not divisible")
```

```
# check given number positive,negative or zero
n= int(input("enter a number"))
if n>0:
  print("+ve")
elif n<0:  
  print("-ve")
else:
  print("zero")
```

```
# given number is leap year or not
n=int(input("enter a year"))
if n%400==0 or (n%100 !=0 and n%4==0):
  print("leap year")
else:
  print("not leap year")  
```
